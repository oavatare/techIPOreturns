```python
# Average S&P price every year

Matrix = [['1990', 334.6320553359684],
 ['1991', 376.18640316205546],
 ['1992', 415.747086614173],
 ['1993', 451.61498023715404],
 ['1994', 460.4159920634921],
 ['1995', 541.7191666666669],
 ['1996', 670.4950787401579],
 ['1997', 873.4278656126485],
 ['1998', 1085.5030158730165],
 ['1999', 1327.329603174603],
 ['2000', 1427.2210714285718],
 ['2001', 1194.1783870967743],
 ['2002', 993.9350793650796],
 ['2003', 965.2272222222225],
 ['2004', 1130.6498809523807],
 ['2005', 1207.2292857142859],
 ['2006', 1310.463904382471],
 ['2007', 1477.1854581673322],
 ['2008', 1220.0409881422931],
 ['2009', 948.0463492063496],
 ['2010', 1139.9657539682544],
 ['2011', 1267.6387698412705],
 ['2012', 1379.3534000000006],
 ['2013', 1643.7989682539685],
 ['2014', 1931.3761111111107],
 ['2015', 2061.067738095238],
 ['2016', 2094.651309523809],
 ['2017', 2449.0763745019913],
 ['2018', 2746.214023904382],
 ['2019', 2763.279375]]
```

```python
w, h = 1, len(ipos);
fiveYearReturns = [[0 for x in range(w)] for y in range(h)] 
tenYearReturns = [[0 for x in range(w)] for y in range(h)] 
```

```python
for i in range(0, len(ipos)):
    fiveYear = int(ipos['Offer Date Plus 5 Years'].iloc[i][-4:])
    tenYear = int(ipos['Offer Date Plus 10 Years'].iloc[i][-4:])
    for j in range(0, len(Matrix)):
        if int(Matrix[j][0]) == fiveYear:
            print(Matrix[j][0])
            print(Matrix[j - 5][0])
            fiveYearReturns[i] = (Matrix[j][1] / Matrix[j - 5][1]) - 1 
            print((Matrix[j][1] / Matrix[j - 5][1]) - 1)
        if int(Matrix[j][0]) == tenYear:
            print(Matrix[j][0])
            print(Matrix[j - 10][0])
            tenYearReturns[i] = (Matrix[j][1] / Matrix[j - 10][1]) - 1 
            print((Matrix[j][1] / Matrix[j - 10][1]) - 1)
print(fiveYearReturns)
print(tenYearReturns)
```

```python
ipos['marketFiveYearPerformance'] = fiveYearReturns
ipos['marketTenYearPerformance'] = tenYearReturns
```

```python
usefulColumns = ['IPO name', 'Offer date', 'Offer Date Plus 5 Years',
       'Offer Date Plus 10 Years', 'Upside/Downside 5 Years Out', 'Upside/Downside 10 Years Out',
       'marketFiveYearPerformance', 'marketTenYearPerformance']
```

```python
filterIPOs = ipos[usefulColumns]
filterIPOs = filterIPOs.iloc[:-1]
filterIPOs
```

```python
averageFiveYearReturnIPO = 0 

for i in range(0, len(filterIPOs)):
    averageFiveYearReturnIPO = averageFiveYearReturnIPO + filterIPOs['Upside/Downside 5 Years Out'].iloc[i]
print(averageFiveYearReturnIPO / len(filterIPOs))
```

```python
averageFiveYearReturnMarket = 0 

for i in range(0, len(filterIPOs)):
    averageFiveYearReturnMarket = averageFiveYearReturnMarket + filterIPOs['marketFiveYearPerformance'].iloc[i]
print(averageFiveYearReturnMarket / len(filterIPOs))
```

```python
averageTenYearReturnIPO = 0 

for i in range(0, len(filterIPOs)):
    averageTenYearReturnIPO = averageTenYearReturnIPO + filterIPOs['Upside/Downside 10 Years Out'].iloc[i]
print(averageTenYearReturnIPO / len(filterIPOs))
```

```python
averageTenYearReturnMarket = 0 

for i in range(0, len(filterIPOs)):
    averageTenYearReturnMarket = averageTenYearReturnMarket + filterIPOs['marketTenYearPerformance'].iloc[i]
print(averageTenYearReturnMarket / len(filterIPOs))
```

```python
w, h = 1, len(filterIPOs);
boolean5YearCheck = [[0 for x in range(w)] for y in range(h)] 
boolean10YearCheck = [[0 for x in range(w)] for y in range(h)] 
```

```python
for i in range(0, len(filterIPOs)):
    if filterIPOs['Upside/Downside 5 Years Out'].iloc[i] >= filterIPOs['marketFiveYearPerformance'].iloc[i]:
        boolean5YearCheck[i] = 1
    else:
        boolean5YearCheck[i] = 0
    if filterIPOs['Upside/Downside 10 Years Out'].iloc[i] >= filterIPOs['marketTenYearPerformance'].iloc[i]:
        boolean10YearCheck[i] = 1
    else:
        boolean10YearCheck[i] = 0
```

```python
filterIPOs['fiveYearPerformanceCheck'] = boolean5YearCheck
filterIPOs['tenYearPerformanceCheck'] = boolean10YearCheck
```

```python
beatFiveYear = 0
beatTenYear = 0

for i in range(0, len(filterIPOs)):
    if filterIPOs['fiveYearPerformanceCheck'].iloc[i] == 1:
        beatFiveYear = beatFiveYear + 1
    if filterIPOs['tenYearPerformanceCheck'].iloc[i] == 1:
        beatTenYear = beatTenYear + 1
print("Out of the", len(filterIPOs), "stocks in this dataset", beatFiveYear / len(filterIPOs), "beat the market over a five")
print("year period, and ", beatTenYear / len(filterIPOs), "beat the market over a ten year period")
```

```python
# Check the actual returns from the IPO against the market over a five or ten year period

filterIPOs['IPOvsMarket5Years'] = filterIPOs['Upside/Downside 5 Years Out'] - filterIPOs['marketFiveYearPerformance'] 
filterIPOs['IPOvsMarket10Years'] = filterIPOs['Upside/Downside 10 Years Out'] - filterIPOs['marketTenYearPerformance']
filterIPOs
```

```python
print("Over an average five year period between 1995 and 2012, tech IPOs on average performed around", 100 * (filterIPOs['Upside/Downside 5 Years Out'].mean() - filterIPOs['marketFiveYearPerformance'].mean()), '% better than the market on average')
print("Over an average ten year period between 2000 and 2017, tech IPOs on average performed around", 100 * (filterIPOs['Upside/Downside 10 Years Out'].mean() - filterIPOs['marketTenYearPerformance'].mean()), '% better than the market on average')
```

```python
# This leads to a paradoxical situation in which most stocks don't beat the market, but the ones that do are such massive 
# outliers especially paradigm shift enabler such as internet stocks, that a portfolio comprised of internet stocks
# purchased between 1990 and 2007 would have beaten the market by a huge amount, but this is mostly due to a few select 
# companies, such as Cisco, Amazon, Baidu, Netflix, Verisign, and F5 Networks, who have been spectacularly successful 
# internet companies.
```

```python
# These returns are incredibly susceptible to top heavy driven returns. As we've already seen, the average IPO didn't
# outperform the market, and when we eliminate the top three, five, and ten performing IPOs over the first five years
# after IPO, we get completely different results
```

```python
yearReturns = pd.DataFrame(Matrix) 

currentSP = float(2826.06)
```

```python
compareReturns = []
spReturns = []

for i in range(0, len(newipos)):    
    #print(newipos['Year'].iloc[i])
    exampleVal = yearReturns[yearReturns[0] == str((newipos['Year'].iloc[i]))]
    #print(exampleVal)
    priorSP = float(exampleVal[1])
    #print(priorSP)
    marketReturn = currentSP / priorSP - 1
    spReturns.append(currentSP / float(exampleVal[1]) - 1)
    compareReturns.append(newipos['Upside'].iloc[i] - (currentSP / float(exampleVal[1]) - 1))
```
